```python
import tensorflow as tf
import os
from google.colab import drive

# Drive the mount
drive.mount('/drive')
print("Drive mounted!")

# Load the model from drive
model = tf.keras.models.load_model('/drive/MyDrive/detectderm_3class.h5')
print(" Model loaded!")

# TFLite convert
converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
tflite_model = converter.convert()
print("Converted!")

# For save
with open('/content/detectderm_offline.tflite', 'wb') as f:
    f.write(tflite_model)

# Size check
h5_size = os.path.getsize('/drive/MyDrive/detectderm_3class.h5') / (1024*1024)
tflite_size = os.path.getsize('/content/detectderm_offline.tflite') / (1024*1024)
print(f".h5 size:     {h5_size:.2f} MB")
print(f".tflite size: {tflite_size:.2f} MB")

# Save on the Drive
import shutil
shutil.copy('/content/detectderm_offline.tflite',
            '/drive/MyDrive/detectderm_offline.tflite')
print("Saved on the Drive")

```
